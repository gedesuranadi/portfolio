```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset/Combined_Flights_2021.csv')

df.head()
```

```python
col = ['FlightDate', 'Airline', 'Origin', 'Dest', 'Cancelled', 'Diverted',
       'CRSDepTime', 'DepTime', 'DepDelay', 'ArrTime',
       'AirTime', 'ActualElapsedTime',
       'Distance', 'Year', 'Quarter', 'Month', 'DayofMonth', 'DayOfWeek',
       'Tail_Number', 'OriginAirportID', 'OriginCityName', 'OriginStateName', 'DestAirportID', 'DestCityName', 'DestStateName']
```

```python
df2 = df
```

```python
df2 =df2[col]

df2.head()
```

```python
df2.isnull().sum()
```

```python
# Drop ActualElapsedTime missing value

df2 = df2.dropna(subset=['ActualElapsedTime'])
```

```python
df2.isnull().sum()
```

```python
# Jumlah penerbangan yang dilakukan oleh setiap airline

airlineCount = pd.DataFrame(df2['Airline'].value_counts())
airlineCount = airlineCount.reset_index()
airlineCount.columns = ['Airline', 'Freq']

airlineCount.head()
```

```python


sns.set(rc={'figure.figsize' : (18, 10), 'figure.facecolor':'lightblue', 'axes.facecolor':'lightblue'})

plot = sns.barplot(data=airlineCount, x="Airline", y="Freq")
for bar in plot.patches:
    plot.annotate(format(int(bar.get_height())),
                  (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()),
                  ha='center', 
                  va='center',
                  #size=15,
                  xytext=(0, 8),
                  textcoords='offset points')

plt.xlabel("Airline", size=14)
plt.ylabel("Frequency", size=14)
plt.title("Flight", size = 18)
plt.xticks(rotation=90)
plt.show()
```

```python

explode = [0.1] * 22

palette_color = sns.color_palette("viridis")
  
sns.set(rc={'figure.figsize' : (10, 10)})

plt.pie(airlineCount['Freq'], labels=airlineCount['Airline'], colors=palette_color, autopct='%.0f%%', explode=explode)  # type: ignore
  

plt.show()
```

```python
# Jumlah kota asal

OriginCityNameCount = pd.DataFrame(df2['OriginCityName'].value_counts())
OriginCityNameCount = OriginCityNameCount.reset_index()
OriginCityNameCount.columns = ['OriginCityName', 'Freq']

display(OriginCityNameCount.head())
print(len(OriginCityNameCount))
```

```python

sns.set(rc={'figure.figsize' : (12, 8), 'figure.facecolor':'lightblue', 'axes.facecolor':'lightblue'})

OriginCityNameCount10 = OriginCityNameCount[0:10]

plot = sns.barplot(data=OriginCityNameCount10, x="OriginCityName", y="Freq")
for bar in plot.patches:
    plot.annotate(format(int(bar.get_height())),
                  (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()),
                  ha='center', 
                  va='center',
                  #size=15,
                  xytext=(0, 8),
                  textcoords='offset points')

plt.xlabel("Name of City", size=14)
plt.ylabel("Frequency", size=14)
plt.title("Most Popular Origin City", size = 18)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set(rc={'figure.figsize' : (12, 8), 'figure.facecolor':'lightblue', 'axes.facecolor':'lightblue'})

OriginCityNameCount10Tail = OriginCityNameCount.tail(10)

plot = sns.barplot(data=OriginCityNameCount10Tail, x="OriginCityName", y="Freq")
for bar in plot.patches:
    plot.annotate(format(int(bar.get_height())),
                  (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()),
                  ha='center', 
                  va='center',
                  #size=15,
                  xytext=(0, 8),
                  textcoords='offset points')

plt.xlabel("Name of City", size=14)
plt.ylabel("Frequency", size=14)
plt.title("Less Popular Origin City", size = 18)
plt.xticks(rotation=90)
plt.show()
```

```python
# Jumlah penerbangan setiap bulan

monthCount = pd.DataFrame(df2['Month'].value_counts())
monthCount = monthCount.reset_index()
monthCount.columns = ['Month', 'Freq']
monthCount = monthCount.sort_values(by = 'Month')
monthName =  ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthCount['monthName'] = monthName
display(monthCount.head(10))
```

```python

fig, ax = plt.subplots(figsize=(12, 8))

sns.lineplot(data=monthCount, x="monthName", y="Freq")

plt.xlabel("Month", size=14)
plt.ylabel("Frequency", size=14)
plt.title("Flight by Month", size = 18)
plt.xticks(rotation=45)
plt.show()
```

```python
airlineList10 =['Southwest Airlines Co.',
 'Delta Air Lines Inc.',
 'SkyWest Airlines Inc.',
 'American Airlines Inc.',
 'United Air Lines Inc.',
 'Republic Airlines',
 'Endeavor Air Inc.',
 'Envoy Air',
 'Comair Inc.',
 'JetBlue Airways']
```

```python
SAC = df2[df2.Airline == 'Southwest Airlines Co.']
DALI = df2[df2.Airline == 'Delta Air Lines Inc.']
SAI = df2[df2.Airline == 'SkyWest Airlines Inc.']
AAI = df2[df2.Airline == 'American Airlines Inc.']
UALI = df2[df2.Airline == 'United Air Lines Inc.']
RA = df2[df2.Airline == 'Republic Airlines']
EAI = df2[df2.Airline == 'Endeavor Air Inc.']
EA = df2[df2.Airline ==  'Envoy Air']
CI = df2[df2.Airline == 'Comair Inc.']
JA = df2[df2.Airline == 'JetBlue Airways']

```

```python
topAirline = SAC.append([DALI, SAI, AAI, UALI, RA, EAI, EA, CI, JA], ignore_index = True)
```

```python
temp = topAirline.groupby(['Airline', 'OriginCityName'])['FlightDate'].count()
temp = pd.DataFrame(temp)
temp = temp.reset_index()
temp.columns = ['Airline', 'Origin', 'Freq']

temp.head()
```

```python
x = temp.sort_values(by='Freq', ascending = False).head(10)
```

```python
temp = df2.groupby(['Month', 'Airline'])['FlightDate'].count()
temp = pd.DataFrame(temp)
temp = temp.reset_index()
temp.columns = ['Month', 'Airline', 'Freq']

temp.head()
```

```python
tempPivot = temp.pivot(index='Month', columns='Airline', values='Freq')  # type: ignore
```

```python
display(tempPivot)
```

```python

sns.set(rc={'figure.figsize' : (16, 10), 'figure.facecolor':'#cff2f3', 'axes.facecolor':'#cff2f3'})

sns.lineplot(data=tempPivot[airlineList10])

plt.xlabel("Month", size=14)
plt.ylabel("Frequency", size=14)
plt.title("Flight by Month of Top 10 Airlines", size = 18)
plt.xticks(rotation=45)
plt.show()
```
